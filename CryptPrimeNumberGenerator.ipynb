### Compare the processing time for two different approaches


```python
print('Prime number generator using denominator up to n//2')
num = PrimeNumber(ran_seed=43)
num.pick_primes(rng=(100_000, 200_000), fn=num.is_prime_a, nr=5)
```

```python
print('Prime number generator using denominator up to the square root of n')
num = PrimeNumber(ran_seed=43)
num.pick_primes(rng=(100_000, 200_000), fn=num.is_prime_b, nr=5)
```

#### For the same results, the time taken for the first function is significantly longer for this range.
